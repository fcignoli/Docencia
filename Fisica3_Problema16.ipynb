```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
#Campos con carga q,q, y -3q 
#Defino campos Ex, Ey, y EZ

def E_x(x, y,z,d): # Tomo k*q=1 para simplificar, 
  r = np.sqrt(x**2+y**2+z**2)
  return -(x/r**3 + (9*d*y*x)/(r**5))

def E_y(x, y,z,d): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2+z**2) 
  return -(((y-3*d)/(r**3)) + (9*d*y*y)/(r**5))

def E_z(x, y,z,d): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2+z**2)
  return -(z/r**3 + (9*d*y*z)/(r**5))

b=1
a=(2/np.sqrt(3))*b


#Aca fijo etntre que valores graficos, va depender si veo de cerca o de lejos.
Xs = np.linspace(-100, 100, 600) #desde lejos
Ys = np.linspace(-100, 100, 600)
X_m, Y_m = np.meshgrid(Xs, Ys)
Ex = E_x(X_m, Y_m,2,1)
Ey = E_y(X_m, Y_m,2,1)

fig, axs = plt.subplots(1,2, figsize=(14, 7))
axs[0].streamplot(Xs, Ys, Ex, Ey, density=1.5)  # Densidad de flechas de campo
axs[0].plot([-a/2], [0], "go", markersize=6) # Pongo puntos representando las cargas
axs[0].plot([a/2], [0], "go", markersize=6)
axs[0].plot([0], [b], "ro", markersize=9)
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Desde lejos')
axs[0].grid(True)


Xs = np.linspace(-3, 3, 200) #desde cerca
Ys = np.linspace(-3, 3, 200)
X_m, Y_m = np.meshgrid(Xs, Ys)
Ex = E_x(X_m, Y_m,2,1)
Ey = E_y(X_m, Y_m,2,1)


axs[1].streamplot(Xs, Ys, Ex, Ey, density=1.5)  
axs[1].plot([-a/2], [0], "go", markersize=6) # Pongo puntos representando las cargas
axs[1].plot([a/2], [0], "go", markersize=6)
axs[1].plot([0], [b], "ro", markersize=9)
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('Desde cerca')
axs[1].grid(True)

fig.tight_layout()
plt.show()
```
